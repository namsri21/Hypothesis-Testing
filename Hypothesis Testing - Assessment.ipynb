#### 1. A physician is evaluating a new diet for her patients with a family history of heart disease. To test the effectiveness of this diet, 16 patients are placed on the diet for 6 months. Their weights and triglyceride levels are measured before and after the study, and the physician wants to know if either set of measurements has changed (Data set: dietstudy.csv) 

###### Hypothesis testing for weight:

H0: weight before = weight after

H1: weight before > weight after

So since there is a before after scenario, we would conduct a paired sample t-test

```python
stats.ttest_rel(a=dataset.wgt0,b=dataset.wgt4)
```

Since we are conducting one tailed test and pvalue/2 <0.05.Thus we reject the null and since sign of statistics is (+), a>b .So we conclude that weight before was greater than weight after with probability=1-(pvalue/2) under confidence level of 95% and type 1 error(alpha)=0.05

###### Hypothesis testing for triglycreide level:

H0: triglycreide before = triglycreide after

H1: triglycreide before > triglycreide after

So since there is a before after scenario, we would conduct a paired sample t-test

```python
stats.ttest_rel(a=dataset.tg0,b=dataset.tg4)
```

Since pvalue >0.05, we do not reject the null hypothesis and conclude that there is no difference between the before and after triglyceride level

#### 2.An analyst at a department store wants to evaluate a recent credit card promotion. To this end, 500 cardholders were randomly selected. Half received an ad promoting a reduced interest rate on purchases made over the next three months, and half received a standard seasonal ad. Is the promotion effective to increase sales? (Data set: creditpromo.csv) 

```python
dataset=pd.read_csv('creditpromo.csv')
```

```python
dataset.sample(10)
```

```python
dataset.value.unique()
```

```python
dataset.info()
```

###### Hypothesis testing for promotion effectiveness:

H0: dollar spend with old promotion = dollar spend with new promotion

H1: dollar spend with old promotion < dollar spend with new promotion

So since there are two different samples that need to be compared(one with old promotion and other with new promotion), we would conduct a two sample t-test (independent)

```python
stats.ttest_ind(a=dataset.dollars[dataset.value==2],
               b=dataset.dollars[dataset.value==1],
               equal_var=False)
```

From the dataset dataset.value==2 indicates "New Promotion" and dataset.value==1 indicates "Standard promotion"

Since we are conducting a one tailed test , we can see that pvalue/2 <0.05 and sign of statistics is +ve a>b i.e dollars spend after new promotion > dollars spend in standard promotion

Hence we conclude that promotion was effective with a probability=1-(pvalue/2) under confidence level of 95% and type 1 error(alpha)=0.05

a. Is the overall population of Seed yield/plant (g) equals to 200? 

###### Hypothesis testing for Seed yield/plant (g) equals to 200:

H0: Seed yield/plant (g) = 200

H1: Seed yield/plant (g) != 200

Since we have population mean available (200) we will conduct one sample t-test


```python
stats.ttest_1samp(a=dataset.Seed_Yield_Plant,
                 popmean= 200)
```

Since it is a two tailed test and pvalue is <0.05, we reject the null

Hence we can conclude that Seed yield/plant (g) != 200 with a probability=1-pvalue under the confidence level of 95% and type 1 error(alpha)=0.05

#### 4.An electronics firm is developing a new DVD player in response to customer requests. Using a prototype, the marketing team has collected focus data for different age groups viz. Under 25; 25-34; 35-44; 45-54; 55-64; 65 and above. Do you think that consumers of various ages rated the design differently? (Data set: dvdplayer.csv). 

```python
dataset=pd.read_csv("dvdplayer.csv")
```

```python
dataset.head(10)
```

###### Hypothesis testing  design rating by various age group:

H0: Rating by all age group are same

H1: Rating by atleast one age group is different

Hence we can use ANOVA here

```python
dataset.Age_group.unique()
```

```python
dataset.info()
```

```python
a=dataset.dvdscore[dataset.Age_group=='65_and_over']
b=dataset.dvdscore[dataset.Age_group=='55_to_64']
c=dataset.dvdscore[dataset.Age_group=='Under_25']
d=dataset.dvdscore[dataset.Age_group=='25_to_34']
e=dataset.dvdscore[dataset.Age_group=='45_to_54']
```

```python
stats.f_oneway(a,b,c,d,e)
```

Since its a two tailed test and pvalue is <0.05,we reject the null

Hence we conclude that atleast one of the age group rated differently with probability=1-pvalue under confidence level 95% and type 1 error(alpha)=0.05

a. Is there any relationship in between labour force status with marital status? 

###### Hypothesis testing  for relation between labour force (categorical data as per data set) and marital status(categorical data as per data set):

HO : labour force and marital status are independent
    
H1 : labour force and marital status are dependent
    
Since both the variables are categorical here , we will do a chi square test to test the relationship

```python
t=pd.crosstab(dataset.wrkstat,dataset.marital,margins=True)
t
```

```python
stats.chi2_contingency(observed=t)
```

Since p value is <0.05 , thus we reject null

Thus we conclude that labour force and marital status have a replationship among them under confidence level of 95% and type 1 error(alpha)=0.05
